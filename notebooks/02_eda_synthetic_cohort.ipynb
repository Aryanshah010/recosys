# Synthetic User Cohort Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append('..')

profiles_df = pd.read_csv('../data/processed/synthetic_user_profiles.csv')
interactions_df = pd.read_csv('../data/processed/synthetic_interactions.csv')
display(profiles_df.head())
```

## Cohort Distribution


```python
cohort_counts = profiles_df['cohort'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=cohort_counts.index, y=cohort_counts.values)
plt.title('Distribution of Synthetic User Archetypes')
plt.ylabel('Number of Users')
plt.xticks(rotation=45)
plt.show()
```

## Interaction Analysis by Cohort


```python
merged_df = interactions_df.merge(profiles_df, on='user_id')
avg_ratings = merged_df.groupby('cohort')['rating'].mean().sort_values()
print('Average Rating by Cohort:')
display(avg_ratings.to_frame())
```
