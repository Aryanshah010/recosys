# Data Cleaning & Validation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

FIG_DIR = '../results/figures'
ratings_df = pd.read_csv('../data/processed/ratings_final.csv')
```

## Sparsity Analysis


```python
num_users = ratings_df['userId'].nunique()
num_movies = ratings_df['movieId'].nunique()
num_ratings = len(ratings_df)

sparsity = 1.0 - (num_ratings / (num_users * num_movies))
print(f'Users: {num_users}')
print(f'Movies: {num_movies}')
print(f'Ratings: {num_ratings}')
print(f'Matrix Sparsity: {sparsity:.4%}')
```

## Interaction Heatmap (Sample)


```python
sample_users = ratings_df['userId'].sample(100, random_state=42)
sample_movies = ratings_df['movieId'].sample(100, random_state=42)

sample_df = ratings_df[(ratings_df['userId'].isin(sample_users)) & (ratings_df['movieId'].isin(sample_movies))]
pivot = sample_df.pivot(index='userId', columns='movieId', values='rating')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot.isnull(), cbar=False, cmap='viridis')
plt.title('Sparsity Heatmap (Yellow = Missing, Purple = Rated)')
plt.savefig(f'{FIG_DIR}/04_sparsity_analysis.png', bbox_inches='tight')
plt.show()
```
